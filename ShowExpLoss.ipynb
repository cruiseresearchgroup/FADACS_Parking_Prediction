```python
import seaborn as sns
import numpy as np
import json
import matplotlib.pyplot as plt
import os

get_color = lambda : "#" + "".join(np.random.choice(list("02468acef"), size=6))

def showAloss(filePath):
    item_list = []

    sns.set(rc={'figure.figsize':(11.7,4.27)})

    #filePath = "./eptest/test2/e7daf4fe-c1be-11ea-b50c-0242ac110004/train_src_mlp_loss.json"
    print(filePath)
    with open(filePath) as json_file:
        data = json.load(json_file)
        for key in data.keys():
            item_list.append(key)
            #print(key)
            sns.tsplot(data[key], range(len(data[key])),color=get_color(),legend=True)

    h = plt.gca().get_lines()
    plt.legend(handles=h, labels=item_list)

    plt.show()
    
def showAExpLosses(path):
    ff = [os.path.join(path, "t1_"+"train_tgt_mlp_loss.json"),
       os.path.join(path, "t1_"+"train_src_mlp_loss.json"),
        os.path.join(path, "t3_"+"train_tgt_mlp_loss.json"),
       os.path.join(path, "t3_"+"train_src_mlp_loss.json"),
        os.path.join(path, "t6_"+"train_tgt_mlp_loss.json"),
        os.path.join(path, "t6_"+"train_src_mlp_loss.json")]
    
    for f in ff:
        if os.path.isfile(f):
            showAloss(f)
            
showAExpLosses("./experiments/FADACS/e7daf4fe-c1be-11ea-b50c-0242ac110004/")
```
